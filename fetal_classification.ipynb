```python
#importing needed modules
try :
    import pandas as pd
    import numpy as np 
except :
    print('something wrong when importing the modules')
```

```python
#load the datasets
path = pd.read_csv('F:\\datasets\\fetal_health.csv') 
df = path.copy()
#make a view of dataset by choosing randomly 
df.sample(10)
```

```python
#getting information from dataset
df.info()
```

```python
#all columns type are numeric so lets get a describtion of dataset
df.describe()
```

```python
#counting null values in each column
df.isnull().sum()
```

```python
#seperate label and drop it from main dataset
label = df['fetal_health']
df_x = df.drop(['fetal_health'],axis=1)
#making sure all work done succesfull
if 'fetal_health' not in df_x.columns:
    print('droping was succesfull')
```

```python
#importing needed modules for split , searching the best params and scale the features
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#scaling the features data frame
total_scaled = pd.DataFrame(scaler.fit_transform(df_x),columns=df_x.columns)
#spliting the features data frame
xtr , xts , ytr , yts = train_test_split(df_x,label,test_size=0.2,random_state=42)
#scaling the splited features
scaled_xtr = pd.DataFrame(scaler.fit_transform(xtr),columns=df_x.columns)
scaled_xts = pd.DataFrame(scaler.fit_transform(xts),columns=df_x.columns)
```

```python
#make view of train features
scaled_xtr.sample(5)
```

```python
#make view of test features
scaled_xts.sample(5)
```

```python
#using K nearest neighbors for predict
from sklearn.neighbors import KNeighborsClassifier
#train the train features
alg = KNeighborsClassifier().fit(xtr,ytr)
#getting accuracy score from train features and label
alg.score(xtr,ytr)*100
```

```python
#getting accuracy score from test features and label
alg.score(xts,yts)*100
#difference is about 5% and it show algorithm work not bad
```

```python
#getting accuracy score from main dataset and label
alg.score(total_scaled,label)*100
```
